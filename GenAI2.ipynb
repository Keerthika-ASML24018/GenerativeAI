```python
import time
import ollama
import pandas as pd

MODEL_NAME = "phi3:mini"

print("Selected model:", MODEL_NAME)
```

```python
try:
    model_information = ollama.show(MODEL_NAME)
    print(f"{MODEL_NAME} is available and ready to use.")

except ollama.ResponseError as error:
    if error.status_code == 404:
        print(f"{MODEL_NAME} is not installed.")
        print("Downloading the model...")
        
        ollama.pull(MODEL_NAME)
        
        print("Model downloaded successfully.")
    else:
        print("Ollama error:", error.error)

except Exception as error:
    print("Unable to connect to Ollama.")
    print("Make sure the Ollama application is running.")
    print("Error:", error)
```

```python
def generate_response(prompt, temperature=0.7):
    if not prompt.strip():
        return {
            "response": "Prompt cannot be empty.",
            "time": 0,
            "word_count": 0
        }

    start_time = time.perf_counter()

    response = ollama.chat(
        model=MODEL_NAME,
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        options={
            "temperature": temperature
        }
    )

    end_time = time.perf_counter()

    generated_text = response["message"]["content"]
    generation_time = end_time - start_time

    return {
        "response": generated_text,
        "time": generation_time,
        "word_count": len(generated_text.split())
    }
```

```python
user_prompt = input("Enter your prompt: ")

result = generate_response(user_prompt)

print("\n===== USER PROMPT =====")
print(user_prompt)

print("\n===== GENERATED RESPONSE =====")
print(result["response"])

print("\n===== RESPONSE DETAILS =====")
print(f"Generation time: {result['time']:.2f} seconds")
print(f"Number of words: {result['word_count']}")
```

```python
test_prompts = [
    {
        "category": "Explanation",
        "prompt": "Explain artificial intelligence in simple words."
    },
    {
        "category": "Creative Writing",
        "prompt": "Write a short story about a robot attending college."
    },
    {
        "category": "Programming",
        "prompt": "Write a Python function to check whether a number is prime."
    }
]

test_results = []

for test in test_prompts:
    print(f"Generating response for: {test['category']}...")

    result = generate_response(test["prompt"])

    test_results.append({
        "Category": test["category"],
        "Prompt": test["prompt"],
        "Response": result["response"],
        "Word Count": result["word_count"],
        "Generation Time": round(result["time"], 2)
    })

print("\nAll prompts tested successfully.")
```

```python
for number, result in enumerate(test_results, start=1):
    print("=" * 80)
    print(f"TEST {number}: {result['Category']}")
    print("=" * 80)

    print("\nPrompt:")
    print(result["Prompt"])

    print("\nGenerated Response:")
    print(result["Response"])

    print("\nWord Count:", result["Word Count"])
    print("Generation Time:", result["Generation Time"], "seconds")
    print()
```

```python
comparison_df = pd.DataFrame(test_results)

comparison_df[
    [
        "Category",
        "Prompt",
        "Word Count",
        "Generation Time"
    ]
]
```

```python
longest_response = comparison_df.loc[
    comparison_df["Word Count"].idxmax()
]

fastest_response = comparison_df.loc[
    comparison_df["Generation Time"].idxmin()
]

print("===== COMPARISON RESULT =====")

print(
    "Longest response:",
    longest_response["Category"],
    f"({longest_response['Word Count']} words)"
)

print(
    "Fastest response:",
    fastest_response["Category"],
    f"({fastest_response['Generation Time']} seconds)"
)
```

```python
temperature_prompt = "Write a creative introduction for an AI-powered college."

temperature_values = [0.2, 0.7, 1.0]

temperature_results = []

for temperature in temperature_values:
    result = generate_response(
        temperature_prompt,
        temperature=temperature
    )

    temperature_results.append({
        "Temperature": temperature,
        "Response": result["response"],
        "Word Count": result["word_count"],
        "Generation Time": round(result["time"], 2)
    })
```

```python
for result in temperature_results:
    print("=" * 80)
    print("Temperature:", result["Temperature"])
    print("=" * 80)

    print(result["Response"])

    print("\nWord Count:", result["Word Count"])
    print("Generation Time:", result["Generation Time"], "seconds")
    print()
```
